# 1.절 조건문(if)
```
if 조건: 
    조건1이 참일 경우 수행할 명령어들
elif 조건2:
    조건1이 거짓이고, 조건2가 참일경우 수행할 명령어들
else: 
    모든 조건을 만족하지 않을때 수행할 명령어들
```


````python
score = int(input('score>'))
if score >=90:
    print("good job")
elif score >=60:
    print("cheer up")
else:
    print("fail")
````

````python
score = int(input('score> '))

if 90 <= score <= 100:
    print("A")
elif 80 <= score < 90:
    print("B")
elif 70 <= score < 80:
    print("C")
elif 60 <= score < 70:
    print("D")
elif 0 <= score < 60:
    print("F")
else:
    print("유효하지 않은 점수입니다.")
    
````

````python
# 점수(0~100점)를 입력받아 학점을 출력하시오(A, B, C, D, F, 유효하지 않은 점수)
score = int(input('점수는 >'))
if 90 <= score <= 100 : # (score>=90) & (score<=100):
    grade = 'A'
elif 80 <= score < 90:
    grade = 'B'
elif 70 <= score < 80:
    grade = 'C'
elif 60 <= score < 70:
    grade ='D'
elif 0 <= score < 60:
    grade = 'F'
else:
    grade = '유효하지 않은 점수'
print("입력하신 점수는 {}점이고, {}등급입니다".format(score, grade))
````


## 2.2 while 문
```
while조건:
    조건이 참일 경우 반복할 문장들
else:
    조건이 거짓일 경우 한번 실행할 문장(else구문 선택사항)
```

````python
# 1~ 100까지 누적합 출력
tot = 0
i = 1
while i <=100:
    tot += i
    i += 1
print(tot)
````

````python
# 1~ 100까지 짝수 누적합 출력
tot = 0
i = 1
while i <=100:
    if i%2==0:
        tot += i
        i += 1
print(tot)
````

````python
# 연습문제 두 정수 입력받아, 작은 수 부터 큰수까지 홀수의 합을 구해서 출력

num_1 = int(input('num1>'))
num_2 = int(input("num2>"))

# 시작과 끝 정리
start = min(num_1, num_2)
end = max(num_1, num_2)
tot=0
for i in range(start, end+1):
    if not i%2==0:
        tot +=i
print(tot)

    
    
````

````python
# 숫자 찾기
'''1~100사이의 정수를 맞추는 프로그램을 구현하시오
(프로그램안에 임의의 숫자가 만들어지고 그 숫자를 맞추도록 합니다)
1단계. 입력된 숫자가 틀리면 UP/DOWN 메시지를 출력하고 숫자를 맞추면 축하메세지와 함꼐 종료합니다(맞출 때까지 반복)
'''
import random
# 1단계 : 1~100 사이의 난수 발생
randomnum = random.randint(1, 100)
print("컴퓨터가 발생한 임의의 수:", randomnum)
# 2단계 : 수를 입력받아 틀리면 UP OR DOWN 메세지를 출력, 맞추면 break하는 반복문
while True:
    inputnum = int(input('1~100 사이의 수를 넣으시오:>'))
    if inputnum<1 or inputnum>100:
        print('장난하냐')
    elif randomnum < inputnum:
        print("down")
    elif randomnum > inputnum:
        print("up")
    else:
        print("ㅊㅋㅊㅋ")
        break
print("end")
````

````python
for i in range(1, 10):
    for j in range(2, 10):
        print(f"{j}x{i}={i*j:>3}", end='\t')
    print()
    
````

````python
# 중첩 루프 탈출 하는 방법 1 : 플레그 변수 이용
flag = False
for a in range(0, 3):    # a = 0, 1, 2
    for b in range(1, 3):  # b = 1, 2
        if a == b :
            flag = True
            break # 반복문을 아예 빠져나오기
        print(a, b)
        # break시 빠져나오는 부분
    if flag:
        break
print('done')
````

````python
# 10번 2단계
'''
1~100 사이의 정수를 맞추기 프로그램
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 숫자를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메세지 뿌리고 끝남
2단계 : 숫자를 못 맞추면 fRom~to를 조정, 맞추면 축하메세지 뿌리고 끝남
3단계 : 2단계의 기회를 5회 제한. 5회의 기회가 지나면 실패 메세지 뿌리고 끝남
(2단계와 3단계 연습문제)
'''
import random
# (1) 1~100사이의 난수 발생
fRom, to = 1, 100
inputNum = 0 # 사용자가 맞출 수가 저장될 변수
randomNum = random.randint(fRom, to)
print('컴퓨터가 발생한 임의의 수 :', randomNum)
# (2) 수를 입력받아 틀리면 범위 조정과 UP/DOWN 메세지 출력, 
#    맞추면 축하축하 메세지와 break하는 반복문
while randomNum!=inputNum :
    inputNum = int(input(f"{fRom}~{to}사이의 수를 맞춰 보세요 :"))
    
    if (inputNum<fRom) | (inputNum>to):
        print('수의 범위를 벗어났습니다')
    elif randomNum < inputNum:
        print('DOWN ↓')
        to = inputNum - 1
    elif randomNum > inputNum:
        print('UP ↑')
        fRom = inputNum+1
    else:
        print('축하 축하 맞추셨습니다')
print('끝')
````

````python
# 10번 3단계
'''
1~100 사이의 정수를 맞추기 프로그램
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 숫자를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메세지 뿌리고 끝남
2단계 : 숫자를 못 맞추면 fRom~to를 조정, 맞추면 축하메세지 뿌리고 끝남
3단계 : 2단계의 기회를 5회 제한. 5회의 기회가 지나면 실패 메세지 뿌리고 끝남
(2단계와 3단계 연습문제)
'''
import random
# (1) 1~100사이의 난수 발생
fRom, to = 1, 100
inputNum = 0 # 사용자가 맞출 수가 저장될 변수
randomNum = random.randint(fRom, to)
print('컴퓨터가 발생한 임의의 수 :', randomNum)
# (2) 수를 입력받아 틀리면 범위 조정과 UP/DOWN 메세지 출력, 
#    맞추면 축하축하 메세지와 break하는 반복문
for i in range(5, 0, -1):
    inputNum = int(input(f"{fRom}~{to}사이의 수를 맞춰 보세요(남은기회 {i}번) :"))
    
    if (inputNum<fRom) | (inputNum>to):
        print('수의 범위를 벗어났습니다')
    elif randomNum < inputNum:
        print('DOWN ↓')
        to = inputNum - 1
    elif randomNum > inputNum:
        print('UP ↑')
        fRom = inputNum+1
    else:
        print('축하 축하 맞추셨습니다')
        break
else:
    print('모든 기회를 다 쓰셨습니다. 땡!')
print('끝')
````
