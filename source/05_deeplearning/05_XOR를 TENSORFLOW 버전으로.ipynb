```python
import numpy as np
```

```python
from tensorflow.keras.models import Sequential   # 모델 객체 생성
from tensorflow.keras.layers import Input, Dense  # layer층을 쌓을 때 필요
```

```python
x_data = np.array([[0,0],
                 [0,1],
                 [1,0],
                 [1,1]])
y_data = np.array([[0], [1], [1], [0]])
```

```python
# 3. 모델 구성
model = Sequential()
model.add(Dense(units=10,
               input_dim=2,
               activation='relu'))
model.add(Dense(units=20, activation='relu'))
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
model.summary()
```

```python
# 4 모델 학습과정 설정
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['binary_accuracy'])
```

```python
hist = model.fit(x_data, y_data,
                epochs=600)
```

```python
hist.history.keys()
```

```python
import matplotlib.pyplot as plt
fig, loss_ax = plt.subplots(figsize=(10,5))
loss_ax.plot(hist.history['loss'], 'r', label='loss')
loss_ax.set_xlabel('epochs')
loss_ax.set_ylabel('loss')
loss_ax.legend(loc='center right')
acc_ax = loss_ax.twinx()
acc_ax.plot(hist.history['binary_accuracy'], 'b', label = 'binary_accuracy')
acc_ax.set_ylabel('binary_acciracy')
acc_ax.legend(bbox_to_anchor=(0.99, 0.4))
```

```python
# 6. 학습평가하기
score = model.evaluate(x_data, y_data)
print(score[0])
print(score[1])
```
